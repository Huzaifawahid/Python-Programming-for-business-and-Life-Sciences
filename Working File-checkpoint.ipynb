```python

Store_open_data["Sales"].hist() # continous data always is for histogram
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
daily_sales = Store_open_data.groupby('Day')['Sales'].sum()

# Plot the bar chart
plt.figure(figsize=(10, 6))
daily_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
plt.figure(figsize=(10, 6))
bar_plot = daily_sales.plot(kind='bar', color='skyblue')

# Format y-axis labels to display values in millions
bar_plot.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: '{:.0f}M'.format(x / 1000000)))

plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales (Millions)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
